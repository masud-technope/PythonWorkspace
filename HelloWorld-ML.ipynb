```python
# !pip install sklearn
# !pip install pydot

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
from sklearn import tree
from subprocess import check_call
import pydot
```

```python
vgdata = pd.read_csv("data/vgsales.csv")
vgdata.columns
```

```python
X = vgdata.drop(columns=["Rank","Name","Other_Sales"])
```

```python
refined=X.drop(columns=["Year","NA_Sales","EU_Sales","JP_Sales"])
```

```python
refined.columns
```

```python
input_data=refined.drop(columns=["Genre"])
Y=refined["Genre"]

# categorical data encoding
input_data["Platform"]=input_data["Platform"].astype("category")
input_data["Publisher"]=input_data["Publisher"].astype("category")
input_data["Platform_cat"]=input_data["Platform"].cat.codes
input_data["Publisher_cat"]=input_data["Publisher"].cat.codes

xx=input_data.drop(columns=["Platform","Publisher"])
# print(xx)


# split the data
x_train, x_test, y_train, y_test = train_test_split(xx,Y,test_size=0.1)

model=DecisionTreeClassifier()
model.fit(x_train,y_train)

tree.export_graphviz(model, out_file="music-recommender.dot", 
                     feature_names=["Publisher_cat", "Platform_cat","Global_Sales"],
                    class_names=sorted(y_train.unique()), 
                     label="all", 
                     rounded=True, 
                     filled=True)



# predictions=model.predict([[30, 20, 250],[26, 22, 125]])
# predictions=model.predict(x_test)

# predictions
# score = accuracy_score(predictions,y_test)
# score

# now save the model
# joblib.dump(model, "music-recommender.joblib")




```
